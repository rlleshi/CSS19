```python
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter

plt.rcParams['figure.figsize'] = (15, 5)
```

# 1st Part

```python
# We save the result in a data frame which is the primary pandas data structure.
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df['time_in_seconds'].plot()
```

```python
start_pos = int(0.8 * len(df))
df = df.drop(df.index[start_pos:])

# What is happening here is that the plotting is taking the time value in the y-axis
# x-axis is just a counter
df.time_in_seconds.plot()
```

```python
df = df.drop(df.index[df["time_in_seconds"] == 0])
df['time_in_seconds'].plot()
```

# 2nd Part

```python
f = open("sample.txt", "r").read()
```

```python
# We use a counter dictionary to do this
c = Counter(f).most_common()

# Sort the Counter dictionary by value
# The result is a list of tuples
#c = c.most_common()

print(c)

# This is how we plot a list of tuples
plt.plot(*zip(*c))
plt.xlabel("Characters")
plt.ylabel("Frequency")
# chars will be plotted on x-axis and frequency on y-axis
# if plt.plot(chars) => chars will be plotted on y and random numbers will be generated on x
```

```python
# Remove 20% of the highest frequency words
c = c[int(0.2*len(c)):]

plt.plot(*zip(*c))
plt.xlabel("Characters")
plt.ylabel("Frequency")
```
